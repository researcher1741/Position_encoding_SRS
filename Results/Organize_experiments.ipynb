## Folder ordering

```python
def seed_filter(text):
    if "beau" in text:
        base = "_beau"
    elif "subfashi" in text:
        base = "_subfashi"
    elif "fashi" in text:
        base = "_fashi"
    elif "video" in text:
        base = "_video"
    elif "game" in text:
        base = "_game"
    elif "submen" in text:
        base = "_submen"
    elif "men" in text:
        base = "_men"
    # Define a list of patterns to replace
    patterns = r"_\d\d\d{0,2}" + base

    # Compile the regex patterns
    regex = re.compile(patterns)

    # Apply the regex substitution to subfolder
    text = regex.sub(base, text)
    return text

def remove_nonesense(file):
    file = file.replace("_R=1_MaskU=0_Samp=0_OFinals=1_","")
    file = file.replace("_Reverse=1_MaskU=0_SamplingMode=0_OnlyFinals=1_","")
    return file

folder_clean = lambda x: x.replace("Val_", "").replace("Test_", "").replace("Train_", "").replace(".png", "").replace(".txt", "")

def delete_empty_subfolders():
    # Get the current working directory
    current_dir = os.getcwd()

    # Traverse the directory tree
    for root, dirs, _ in os.walk(current_dir, topdown=False):
        for directory in dirs:
            dir_path = os.path.join(root, directory)
            try:
                # Attempt to remove the directory if it is empty
                os.rmdir(dir_path)
                print(f"Deleted empty folder: {dir_path}")
            except OSError:
                # Ignore the error if the folder is not empty
                pass
    time.sleep(5)
```

### RERANGE

```python
def rearrange_string(input_string, numbers_list = [41, 42, 1741, 1742, 1810, 1809]):
    input_string, dot, extension = input_string.rpartition('.')
    if "beauty" in input_string:
        base = "beauty"
    elif "fashion" in input_string:
        base = "fashion"
    elif "video_games" in input_string:
        base = "video_games"
    elif "men" in input_string:
        base = "men"
    try:
        input_string = input_string.replace(base, "")
    except:
#         print(f'base {base}')
        print(f'input_string {input_string}')
        quit()
        
    # Find all occurrences of numbers in the input string
    occurrences = [(number, input_string.find(str(number))) for number in numbers_list]
    # Remove occurrences with -1 index (not found in string)
    occurrences = [(number, index) for number, index in occurrences if index != -1]
    # Sort occurrences based on their indices
    occurrences.sort(key=lambda x: x[1])

    # Rearrange the string based on occurrences
    result = input_string
    for number, index in occurrences:
        result = result.replace(str(number), '') + '_' + str(number) + '_' + base
    result = result.rstrip('_')
    result = result.replace('__','_').replace('__','_').replace('__','_')
    if base not in result:
        result += "_" + base
    result += "." + extension
    return result
text = "carca_nmax_0.0001_ntype_0.1_1741_video_games.txt"
rearrange_string(text)
```

```python
def find_empty_txt_files(root_dir, delete=False):
    empty_files = []

    for foldername, subfolders, filenames in os.walk(root_dir):
        for filename in filenames:
            if filename.endswith('.txt'):
                filepath = os.path.join(foldername, filename)
                if os.path.getsize(filepath) == 0:
                    empty_files.append(filepath)
                    if delete:
                        os.remove(filepath)

    return empty_files

empty_txt_files = find_empty_txt_files(cwd, delete=False)
print(empty_txt_files)
```
