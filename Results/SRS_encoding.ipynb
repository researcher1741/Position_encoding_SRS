```python
stats = {"Subgames": {
    "name": "Subgames",
    "users": 5000,
    "items": 4601,
    "interactions": 33353,
    "density": r"$1.3341 times 10 3$",
},
"Submen2": { # old submen3
    "name": "Submen2",
    "users": 10000,
    "items": 45129,
    "interactions": 74391,
    "density": r"1.648 times 10 4$",
},
"Submen": {
    "name": "Submen",
    "users": 8332,
    "items": 10000,
    "interactions": 28603,
    "density": r"$2.860 times 10 4$",
},
"Subfashion": {
    "name": "Subfashion",
    "users": 3840,
    "items": 5000,
    "interactions": 16080,
    "density": r"$3.564 times 10 4$",
}}

def density_table(list_df, stats):
    data = pd.DataFrame(columns=["Dataset", "Users", "Items", "Interactions", "Hit Dev.",
                             "NDCG Dev.", "CI-length", "Sparsity"])
    for (name, df1) in list_df:
        # Drop rows with NaN values and filter by 'encoding' == 'carca'
        if "games" not in name:
            df1 = df1.dropna(axis=0)
        df1 = df1[df1['encoding'].isin([r'\None'])]

        # Append the results to the DataFrame
        data = data.append({
            "Dataset": name,
            "Users": stats[name]["users"],
            "Items": stats[name]["items"],
            "Interactions": stats[name]["interactions"],
            "Hit Dev.":  df1['dev_hit'].mean(),
            "NDCG Dev.": df1['dev_ndcg'].mean(),
            "CI-length": df1['CI_length'].mean(),
            "Sparsity": stats[name]["density"],
        }, ignore_index=True)

    # Output the final DataFrame
    print(data.to_latex(index=False))

devhit = lambda df: round(df[df['encoding'].isin([r'\None', r'\RotatoryCon', r'\RMHA'])]["dev_hit"].mean(), 4)
devndcg = lambda df: round(df[df['encoding'].isin([r'\None', r'\RotatoryCon', r'\RMHA'])]["dev_ndcg"].mean(), 4)
whole = lambda df: print(cleanit_for_position(df).to_latex(index=False))
show = lambda df, num: print(best(df[mylist], num).head(4).to_latex(index=False))

def greate_tables(list_df, stats):
    """ Compute the statistics and print them in latex format to be used ta the paper. """
    # This function is split into games and not because of the nan values in the norm maximum.
    list_df = [(df["dataset"][0], df) for df in list_df]
    list_games = [(name, df[df["nmax"].isna()]) for (name, df) in list_df if "game" in name]
    
    for case in [list_df, list_games]:
        print("########### STATISTICS 1")
        for (name, df1) in case:
            print(f" Dataset {name}, Hit dev total: {devhit(df1)}, NDCG dev total: {devndcg(df1)}")
        print("\n")
        print("########### STATISTICS 2")
        density_table(case, stats)
    
        print("########### all the tables")
        for (name, df1) in case:
            print(name)
            whole(df1)
        print("\n")
        print("########### max 3")
        for (name, df1) in case:
            print(name)
            show(df1, 3)
        print("\n")
        print("########### max 12")
        for (name, df1) in case:
            print(name)
            show(df1, 12)
    
```

# P-values

```python
from scipy.stats import ttest_1samp

# Assuming 'data' is a NumPy array or a list containing your performance measurements
reference_value = 0.550  # The value you want to compare against
data = np.array([42.5, 59.53, 43.11, 35.72, 63.72, 34.86])

# Perform one-sample t-test
t_stat, p_value = ttest_1samp(data, reference_value)

# Check if the p-value is less than the significance level (e.g., 0.05)
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. The mean performance is significantly different.")
else:
    print("Fail to reject the null hypothesis. The mean performance is not significantly different.")
p_value
```
